# 03 — German Frequency Lookup

Since grouping is by **category** (handled in notebooks 01/04) rather than frequency
band, it just merges German SUBTLEX-DE frequency values onto the translated selection, so we have Zipf values available for
balance-checking across categories and so notebook 04 has a `ZipfSUBTLEX` column to report alongside the closeness-based German CLIP selection.


```python
import pandas as pd
```

```python
# ===== Load translated words + German frequency norms =====
word_df = pd.read_csv("output/english_to_german_selection.tsv", sep="\t", encoding="utf-8")

subtlex = pd.read_excel(
    "SUBTLEX-DE/SUBTLEX-DE cleaned version with Zipf values.xlsx",
    sheet_name=0,
    engine="openpyxl"
)
```

```python
word_df["word_de"] = word_df["GermanTranslation"].str.lower()
subtlex["orth_lower"] = subtlex["Word"].str.lower()

subtlex_small = subtlex[["orth_lower", "SUBTLEX", "ZipfSUBTLEX"]]
merged = pd.merge(
    word_df, subtlex_small,
    how="left", left_on="word_de", right_on="orth_lower"
)
```

```python
# Words with no German frequency match — have to be checked manually
missing = merged[merged["ZipfSUBTLEX"].isna()]
print(f"{len(missing)} words with no SUBTLEX-DE match:")
print(missing[["Word", "GermanTranslation"]])

merged.to_csv("output/german_selection.csv", index=False, encoding="utf-8-sig")
print("Saved output/german_selection.csv")
merged[["word_de", "Word", "GermanTranslation", "category", "ZipfSUBTLEX"]]
```
